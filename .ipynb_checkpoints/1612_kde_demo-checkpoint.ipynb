```python
import pandas as pd
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
from numpy.random import randn
data = randn(50)
```

```python
data
```

```python
plt.hist(data, alpha=0.5)
sns.rugplot(data)
#データが多いところはヒストグラムが高くなる。
```

```python
#kdeplot
sns.kdeplot(data)
sns.rugplot(data)
```

```python
#バンド幅の違いでグラフを表示
sns.rugplot(data)

list =  [0.1, 0.5, 1.0]
for num in list:
    sns.kdeplot(data, bw=num, label=num)
```
