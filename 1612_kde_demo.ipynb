#カーネル密度推定の基礎的なデモ 

解説はこちらから => http://www.ie-kau.net/entry/kernel_density_est

```python
import pandas as pd
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
%matplotlib inline
sns.set_style('whitegrid')
```

```python
from numpy.random import randn
data = randn(50)
```

```python
data
```

```python
plt.hist(data, alpha=0.5)
sns.rugplot(data)
#データが多いところはヒストグラムが高くなる。
```

```python
#kdeplot
sns.distplot(data)
sns.rugplot(data)
```

```python
#バンド幅の違いでグラフを表示
sns.rugplot(data)
list =  [0.1, 0.3, 0.5]

for num in list:
    sns.kdeplot(data, bw=num, label=num)
```

```python
#カーネル関数の違いでグラフを表示
#https://en.wikipedia.org/wiki/Kernel_(statistics)
sns.rugplot(data)
options = ['biw', 'tri', 'gau']

for option in options:
    sns.kdeplot(data, kernel=option, label=option, bw=0.5)
```

```python
#2次元でのカーネル密度推定
iris = sns.load_dataset("iris")
sns.kdeplot(iris.sepal_length, iris.petal_width, shade=True)
```

```python
#こっちのほうが便利できれい
sns.jointplot(x="sepal_length", y="petal_width", kind="kde", data=iris);
```
